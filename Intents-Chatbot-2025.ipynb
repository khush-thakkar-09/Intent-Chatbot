**Preprocessing function — tokenize, lowercase, remove stopwords/punctuation, lemmatize:- ⬇**

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import string

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')

def preprocess(text):
  tokens = nltk.word_tokenize(text)
  tokens = [w.lower() for w in tokens]
  stop_words = set(stopwords.words('english'))
  tokens = [w for w in tokens if w not in stop_words and w not in string.punctuation]
  lemmatizer = WordNetLemmatizer()
  tokens = [lemmatizer.lemmatize(w) for w in tokens]
  return tokens
```

```python
processed_patterns  = [preprocess(pattern) for pattern in all_patterns]
for i in range(10):
  print(f'Original: {all_patterns[i]}')
  print(f'Processed: {processed_patterns[i]}')
  print()
```

Neural networks need numeric data.

```python
from sklearn.preprocessing import LabelEncoder

training_sentences = [doc[0] for doc in documents]
training_labels = [doc[1] for doc in documents]

lenc = LabelEncoder()
training_labels = lenc.fit_transform(training_labels)
```

**Creation of training Data II (for features/sentences):-** (preprocessing using Tensorflow - converting text to numbers and making sequences of equal length)

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

vocab_size = 1000
embedding_dim = 50
max_len = 20
oov_tok = "<OOV>"

tokenizer = Tokenizer(num_words=vocab_size, oov_token = oov_tok)
tokenizer.fit_on_texts(training_sentences)

sequences = tokenizer.texts_to_sequences(training_sentences)
padded_sequences = pad_sequences(sequences, truncating='post', maxlen=max_len,padding='post')
```

```python
import numpy as np

embedding_dim = 50
embedding_index = {}

with open('glove.6B.50d.txt',encoding='utf-8') as f:
  for line in f:
    values = line.split()
    word = values[0]
    coefs = np.asarray(values[1:], dtype='float32')
    embedding_index[word] = coefs

embedding_matrix = np.zeros((vocab_size,embedding_dim)) #Keras wants the embedding weights in a matrix: row = word, columns = embedding numbers.
for word, i in tokenizer.word_index.items():
    if i < vocab_size:
        embedding_vector = embedding_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
```

```python
words = list(tokenizer.word_index.keys())
word_vectors = []

for w in words:
    if w in embedding_index:
        word_vectors.append(embedding_index[w])
    else:
        word_vectors.append(np.zeros(embedding_dim))

word_vectors = np.array(word_vectors)

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
points = pca.fit_transform(word_vectors)

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
plt.scatter(points[:, 0], points[:, 1], s=10, alpha=0.5)

for i, word in enumerate(words[:30]):
    plt.annotate(word, (points[i, 0], points[i, 1]), fontsize=10)

plt.title("PCA Visualization of All Vocabulary Words in GloVe Space")
plt.xlabel("Component 1")
plt.ylabel("Component 2")
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=0, perplexity=30)
points_tsne = tsne.fit_transform(word_vectors)

plt.figure(figsize=(12, 8))
plt.scatter(points_tsne[:, 0], points_tsne[:, 1], s=10, alpha=0.5)
for i, word in enumerate(words[:30]):
    plt.annotate(word, (points_tsne[i, 0], points_tsne[i, 1]), fontsize=10)
plt.title("t-SNE Visualization of All Vocabulary Words in GloVe Space")
plt.show()
```

**Splitting Data:-**

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(padded_sequences,training_labels,test_size=0.2,random_state=42,stratify=training_labels)
```

**Creating a model:-**

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Bidirectional, LSTM, Dropout, Dense

embedding_layer = Embedding(
    input_dim=vocab_size,
    output_dim=embedding_dim,
    weights=[embedding_matrix],
    input_length=max_len,
    trainable=True  # (Try True later for fine-tuning)
)

model = Sequential([
    embedding_layer,
    Bidirectional(LSTM(64)),   # Bidirectional for richer context, 32 units is “small”
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dense(len(set(training_labels)), activation='softmax')
])

model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

```

```python
from tensorflow.keras.callbacks import EarlyStopping

es = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True,min_delta=0.00001,mode='min')

history = model.fit(padded_sequences, training_labels, epochs=100,batch_size=16,validation_data=(X_val,y_val),verbose=1,callbacks=[es])
```

```python
import matplotlib.pyplot as plt

# Accuracy
plt.plot(history.history['accuracy'], label='train_acc')
plt.plot(history.history['val_accuracy'], label='val_acc')
plt.title('Accuracy')
plt.legend()
plt.show()

# Loss
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title('Loss')
plt.legend()
plt.show()
```

```python
def model_predict(user_input,model,tokenizer,lbl_encoder,max_len):
  padded_seq = preprocess_user_input(user_input,tokenizer,max_len)
  preds = model.predict(padded_seq)
  pred_class = preds.argmax(axis=1)[0]
  intents = lbl_encoder.inverse_transform([pred_class])[0]
  return intents
```

```python
import random
def get_responses(intents,data):
  for i in data['intents']:
    if i['tag'] == intent:
      return random.choice(i['responses'])
  return "Sorry, I don't understand..."
```
